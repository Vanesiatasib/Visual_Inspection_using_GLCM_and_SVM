```python
import os
import cv2

# Folder containing images to process
ng_folder = r"D:\Proposal Skripsi gas 2024\skripsi\bahan_penelitian\ng"
ng_list = os.listdir(ng_folder)

# Coordinates for cropping
x, y, w, h = 576, 446, 95, 174

# Output folder to save cropped images
output_folder = r"D:\Proposal Skripsi gas 2024\skripsi\bahan_penelitian\ng_crop"
os.makedirs(output_folder, exist_ok=True)

# Process each image in the ng_folder
for ng in ng_list:
    img_path = os.path.join(ng_folder, ng)
    img = cv2.imread(img_path)
    
    if img is not None:  # Ensure the image was loaded successfully
        # Crop region of interest (ROI)
        roi_crop = img[y:y+h, x:x+w]
        
        # Define output path for each image
        output_path = os.path.join(output_folder, f"crop_{ng}")
        cv2.imwrite(output_path, roi_crop)
        print("Hasil crop disimpan di:", output_path)
    else:
        print(f"Could not load image: {img_path}")
```

```python
import cv2
import os

ref_point = []
cropping = False

def select_roi(event, x, y, flags, param):
    global ref_point, cropping

    if event == cv2.EVENT_LBUTTONDOWN:
        ref_point = [(x, y)]
        cropping = True

    elif event == cv2.EVENT_LBUTTONUP:
        ref_point.append((x, y))
        cropping = False
        cv2.rectangle(image, ref_point[0], ref_point[1], (0, 255, 0), 2)
        cv2.imshow("image", image)

image_path = r"D:\Proposal Skripsi gas 2024\skripsi\bahan_penelitian\ng\NG-1-20241007-112615-933.BMP"

if not os.path.exists(image_path):
    print(f"File tidak ditemukan di lokasi: {image_path}")
    exit()

image = cv2.imread(image_path)
if image is None:
    print("Gambar tidak ditemukan. Periksa kembali path gambar atau format file.")
    exit()

clone = image.copy()

cv2.namedWindow("image")
cv2.setMouseCallback("image", select_roi)

while True:
    # Tampilkan gambar
    cv2.imshow("image", image)
    key = cv2.waitKey(1) & 0xFF

    if key == ord("r"):
        image = clone.copy()

    elif key == ord("c"):
        break

if len(ref_point) == 2:
    x1, y1 = ref_point[0]
    x2, y2 = ref_point[1]

    if 0 <= x1 < x2 <= clone.shape[1] and 0 <= y1 < y2 <= clone.shape[0]:
        roi = clone[y1:y2, x1:x2]
        cv2.imshow("Cropped Image", roi)
        cv2.imwrite("cropped_image.jpg", roi)
        cv2.waitKey(0)
    else:
        print("Koordinat cropping berada di luar batas gambar.")
cv2.destroyAllWindows()

if len(ref_point) == 2:
    print(f"Koordinat awal (x1, y1): {ref_point[0]}")
    print(f"Koordinat akhir (x2, y2): {ref_point[1]}")

    width = x2 - x1
    height = y2 - y1
    print(f"Lebar ROI: {width}")
    print(f"Tinggi ROI: {height}")
```
